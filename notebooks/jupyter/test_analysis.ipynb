```python
import sys
from pathlib import Path

project_root = str(Path.cwd().parent.parent)
if project_root not in sys.path:
    sys.path.insert(0, project_root)
```

```python
from src.helpers import calculate_growth_rate  # noqa: E402
```

```python
# Test positive growth
result = calculate_growth_rate(110, 100)
assert abs(result - 10.0) < 1e-9, f"Expected 10.0, got {result}"
print("PASSED: positive growth")
```

```python
# Test negative growth
result = calculate_growth_rate(90, 100)
assert abs(result - (-10.0)) < 1e-9, f"Expected -10.0, got {result}"
print("PASSED: negative growth")
```

```python
# Test zero previous returns None
result = calculate_growth_rate(100, 0)
assert result is None, f"Expected None, got {result}"
print("PASSED: zero previous")
```

```python
# Test no change
result = calculate_growth_rate(100, 100)
assert abs(result - 0.0) < 1e-9, f"Expected 0.0, got {result}"
print("PASSED: no change")
```
